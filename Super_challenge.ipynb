```python
import pandas as pd
```

```python
from io import StringIO
```

```python
data = StringIO('Airline Code;DelayTimes;FlightCodes;To_From\nAir Canada (!);[21, 40];20015.0;WAterLoo_NEWYork\n<Air France> (12);[];;Montreal_TORONTO\n(Porter Airways. );[60, 22, 87];20035.0;CALgary_Ottawa\n12. Air France;[78, 66];;Ottawa_VANcouvER\n""".\\.Lufthansa.\\.""";[12, 33];20055.0;london_MONTreal\n'

)
```

```python
df =  pd.read_csv(data, delimiter = ';', lineterminator = '\n')
```

```python
df
```

```python
#1.0 FlightCodes column
df['FlightCodes'].fillna(method = 'ffill', inplace = True)
df['FlightCodes'] = df['FlightCodes'].astype(int)
```

```python
#1.0 FlightCodes column
df['FlightCodes'] = df['FlightCodes'].astype(int)
for i in range(1,len(df.axes[1])):
  if df['FlightCodes'][i] == df['FlightCodes'][i-1]:
    df['FlightCodes'][i] = df['FlightCodes'][i-1] +10
```

```python
df
```

```python
#2.0 To_From Column
df[['To', 'From']] = df['To_From'].str.split('_', expand =True)
```

```python
#2.0
df.drop(columns = ['To_From'], inplace = True)
```

```python
df
```

```python
#2.0
df['To'] = df['To'].str.upper()
df['From'] = df['From'].str.upper()
```

```python
df
```

```python
#3.0 Airline Code Column
df['Airline Code'] = df['Airline Code'].str.replace('[^\w\s]','')
```
